```python
kub1 = [1, 3, 4, 5, 6, 8]
kub2 = [1, 2, 2, 3, 3, 4]
kub = [1, 2, 3, 4, 5, 6]
def probability(kub1, kub2):
    for i in range(2, 13):
        s = 0
        for j in range(6):
            for k in range(6):
                if kub1[j] + kub2[k] == i:
                    s += 1
        print('Вероятность выпадения суммы', str(i) + ':', str(round(s*100/36, 2)) + '%')
probability(kub1, kub2)
print()
probability(kub, kub)
#Вероятности выпадения каждой суммы из значений данных игральных костей равны вероятностям для стандартной пары игральных костей.
```
